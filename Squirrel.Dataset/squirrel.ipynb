```python
# Replace values in 'sex' column
df['sex'] = df['sex'].replace({0: 'Female', 1: 'Male'})

# Convert 'diet' column to categorical type with categories 'A', 'B', 'C'
df['diet'] = df['diet'].replace({1: 'A', 2: 'B', 3: 'C'})
df['diet'] = pd.Categorical(df['diet'], categories=['A', 'B', 'C'], ordered=True)

# Create plot
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(data=df, x='diet', y='weight', hue='sex', palette={'Female': 'magenta', 'Male': 'green'})

# Add legend
scatter.legend(title='Sex')

# Set title and labels
plt.title("Red Squirrel Diet Study")
plt.xlabel("Diet Label")
plt.ylabel("Adult Weight/lbs")

# Show plot
plt.show()
```
